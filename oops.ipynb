```python
class student:
   def __init__(self,name,age,grade):
     self.name=name
     self.age=age
     self.grade=grade
   def display(self):
     print("name",self.name)
     print("age",self.age)
     print("grade",self.grade)
s1 = student("palak",18,"A+")
s1.display()
```

inheritance in oops

```python
class parent:
  def test_parent(self):
    print("class")
```

```python
class child(parent):
  pass
```

```python
child_obj=child()
```

```python
child_obj.test_parent()
```

```python
class class1:
  def test_class1(self):
    print("class1")
```

```python
class class2(class1):
  def test_class2(self):
    print("class2")
```

```python
class class3(class2):
  def test_class3(self):
    print("class3")
```

```python
obj_class3 = class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class2 = class2()
```

```python
obj_class3.test_class2()
```
